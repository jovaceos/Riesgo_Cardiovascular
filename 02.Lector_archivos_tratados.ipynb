```python
# Ruta de la carpeta en Google Drive que contiene los archivos XLSX
carpeta_drive = '/content/drive/Proyecto_Aplicado/Bases_Tratadas'
```

```python
# Lista para almacenar los resultados
informe_general = []

# Obtener la lista de archivos XLSX en la carpeta
archivos = [archivo for archivo in os.listdir(carpeta_drive) if archivo.endswith('.xlsx')]

# Función para explorar cada archivo XLSX
def explorar_archivo(archivo):
    ruta_archivo = os.path.join(carpeta_drive, archivo)
    datos = pd.read_excel(ruta_archivo)

    # Obtener información del archivo
    informacion = datos.info()
    cantidad_datos_columna = datos.shape [0]
    cantidad_columnas = datos.shape[1]
    cantidad_total_datos = datos.shape[0] * datos.shape[1]
    cantidad_total_nulos = datos.isnull().sum().sum()
    porcentaje_total_nulos = (cantidad_total_nulos / cantidad_total_datos) * 100

    # Agregar resultados al informe
    informe_general.append({
        'Archivo': archivo,
        'Cantidad de datos por columna': cantidad_datos_columna,
        'Cantidad de columnas': cantidad_columnas,
        'Cantidad Total de Datos': cantidad_total_datos,
        'Cantidad Total de Nulos': cantidad_total_nulos,
        'Porcentaje de Nulos': porcentaje_total_nulos,
        })

# Iterar sobre cada archivo XLSX
for archivo in archivos:
    explorar_archivo(archivo)

# Crear DataFrame a partir del informe
informe_general_df = pd.DataFrame(informe_general)

# Mostrar el informe
for info in informe_general:
    print(info)

# Guardar el informe como un archivo CSV
informe_general_df.to_csv('/content/drive/Proyecto_Aplicado/Informe_Archivos_Bases_Tratadas/informe_general.csv', index=False)

# Mostrar el informe
print(informe_general)
```

```python
# Función para obtener información por columna
# Lista para almacenar los resultados
informe_columnas_texto = []
informe_columnas_numericas = []
informe_columnas_fecha = []

# Obtener la lista de archivos XLSX en la carpeta
archivos = [archivo for archivo in os.listdir(carpeta_drive) if archivo.endswith('.xlsx')]

# Función para explorar cada archivo XLSX
def explorar_archivo(archivo):
    ruta_archivo = os.path.join(carpeta_drive, archivo)
    datos = pd.read_excel(ruta_archivo)

    for columna in datos.columns:
      tipo_columna = datos[columna].dtype
      if pd.api.types.is_object_dtype(tipo_columna):
        # Columna de texto
        categorias = datos [columna]. value_counts()
        porcentaje_categorias = 100* (datos [columna]. value_counts()/len(datos [columna]))
        cantidad_nulos_columna = datos[columna].isnull().sum()
        valores_unicos = datos[columna].nunique()
        informe_columnas_texto.append({
            'Archivo': archivo,
            'Columna': columna,
            'Tipo': 'Texto',
            'Categorias': categorias,
            'Porcentaje categorias': porcentaje_categorias,
            'Cantidad de Nulos': cantidad_nulos_columna,
            'Valores Únicos': valores_unicos,
            })
      elif pd.api.types.is_float_dtype(tipo_columna):
          # columna números
          cantidad_nulos_columna = datos[columna].isnull().sum()
          descripcion = datos[columna].describe()
          informe_columnas_numericas.append({
              'Archivo': archivo,
              'Columna': columna,
              'Tipo': 'Numérico',
              'Cantidad de Nulos': cantidad_nulos_columna,
              'Estadísticos': descripcion
              #'Promedio': promedio,
              #'Desviación Estándar': desviacion_estandar
            })
      elif pd.api.types.is_datetime64_any_dtype(tipo_columna):
            # Columna de fecha
            cantidad_nulos_columna = datos[columna].isnull().sum()
            fecha_minima = datos[columna].min()
            fecha_maxima = datos[columna].max()
            informe_columnas_fecha.append({
                'Archivo': archivo,
                'Columna': columna,
                'Tipo': 'Fecha',
                'Cantidad de Nulos': cantidad_nulos_columna,
                'Fecha Mínima': fecha_minima,
                'Fecha Máxima': fecha_maxima
            })

# Procesar cada archivo
for archivo in archivos:
  explorar_archivo(archivo)

# Crear DataFrame a partir del informe
informe_texto_df = pd.DataFrame(informe_columnas_texto)
informe_numerico_df = pd.DataFrame(informe_columnas_numericas)
informe_fechas_df = pd.DataFrame(informe_columnas_fecha)

# Guardar el informe como un archivo CSV
informe_texto_df.to_csv('https://drive.google.com/drive/folders/1eNhb4lY8wwjeGTnPTLV0ZEnWCTp0d580?usp=drive_link/informe_coltexto.csv', index=False)
informe_numerico_df.to_csv('https://drive.google.com/drive/folders/1eNhb4lY8wwjeGTnPTLV0ZEnWCTp0d580?usp=drive_link/informe_colnum.csv', index=False)
informe_fechas_df.to_csv('https://drive.google.com/drive/folders/1eNhb4lY8wwjeGTnPTLV0ZEnWCTp0d580?usp=drive_link/informe_colfecha.csv', index=False)

# Mostrar el informe
for info in informe_columnas_texto, informe_columnas_numericas, informe_columnas_fecha:
    print(info)
```
