```python
datos = '/content/drive/MyDrive/MCD/Proyecto_Aplicado/Tabla_maestra/tabla_maestra_analisis_exploratorio.csv'
```

```python
datos_1 = pd.read_csv(datos)
datos_1
```

```python
# Obtener los nombres de las columnas
columnas = datos_1.columns

# Obtener la cantidad de nulos por columna
nulos_por_columna = datos_1.isnull().sum()

# Obtener el tipo de datos de cada columna
tipos_de_datos = datos_1.dtypes

#Obtener el porcentaje de nulidad
nulidad = datos_1.isnull().sum()/len(datos_1)

# Combinar la información en un DataFrame
informacion_columnas = pd.DataFrame({
    'Nombre de la Columna': columnas,
    'Cantidad de Nulos': nulos_por_columna,
    'Tipo de Datos': tipos_de_datos,
    'Nulidad': nulidad
})

print("\nInformación de las columnas:")
print(informacion_columnas)
```

```python
caracterizacion = informacion_columnas
```

```python
caracterizacion
```

```python
caracterizacion.to_csv('caracterizacion.csv', index=False)
```

```python
# Crear el histograma
plt.figure(figsize=(8, 6))
plt.hist(caracterizacion['Nulidad'], bins=10, color='gray', edgecolor='black')

# Añadir etiquetas y título
plt.xlabel('% de nulidad')
plt.ylabel('Frecuencia')
plt.title('Nulidad de las 104 variables')
plt.grid()

# Mostrar el gráfico
plt.show()
```
