```python
# Directorio donde se encuentran los archivos de Excel
directorio = '/content/drive/MyDrive/MCD/Proyecto_Aplicado/Bases_tratadas'
```

```python
# Crear una lista vacía en la que se almacenan los dataframes
dfs = []

# iterate over the excel files
# Recorre todos los archivos en el directorio
for archivo in os.listdir(directorio):
    if archivo.endswith('.xlsx'):  # Asegúrate de que sean archivos Excel
        # Lee el archivo Excel
        df = pd.read_excel(os.path.join(directorio, archivo))
        # Agregar nueva columna con el nombre del archivo
        df.insert(0, 'File Name', os.path.basename(archivo))
        # Agregar el dataframe a la lista
        dfs.append(df)
```

```python
# Concatenar los dataframes en la lista en un único dataframe
df = pd.concat(dfs)

# Guardar el dataframe en un archivo excel
df.to_csv('/content/drive/MyDrive/MCD/Proyecto_Aplicado/Tabla_maestra/tabla_maestra.csv', index=False)
```
