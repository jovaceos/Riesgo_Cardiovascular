# GRÁFICOS

```python
frecuencias = tabla4['CLASIFICACION FINAL DEL RIESGO'].value_counts()

# Crear el gráfico de torta
plt.figure(figsize=(5, 5))
plt.pie(frecuencias, labels=frecuencias.index, autopct='%1.1f%%', startangle=140)
plt.title('Distribución de la clasificación del riesgo')
plt.show()
```

```python
frecuencias = tabla4['GRUPO EDAD'].value_counts()

# Crear el gráfico de torta
plt.figure(figsize=(5, 5))
plt.pie(frecuencias, labels=frecuencias.index, autopct='%1.1f%%', startangle=140)
plt.title('Distribución del grupo de edad')
plt.show()
```

```python
frecuencias = tabla4['8_Sexo'].value_counts()

# Crear el gráfico de torta
plt.figure(figsize=(5, 5))
plt.pie(frecuencias, labels=frecuencias.index, autopct='%1.1f%%', startangle=140)
plt.title('Distribución de genero')
plt.show()
```

```python
# Filtrar el DataFrame donde la columna C2 es 'M'
df_filtrado = tabla4[tabla4['8_Sexo'] == 'M']

# Contar las frecuencias de cada categoría en la columna C1 para el DataFrame filtrado
frecuencias = df_filtrado['CLASIFICACION FINAL DEL RIESGO'].value_counts()

# Crear el gráfico de torta
plt.figure(figsize=(8, 8))
wedges, texts, autotexts = plt.pie(
    frecuencias,
    labels=frecuencias.index,
    autopct='%1.1f%%',
    startangle=140,
    colors=plt.cm.Paired(range(len(frecuencias)))
)

# Añadir una leyenda
plt.legend(wedges, frecuencias.index, title="Riesgo", loc="center left", bbox_to_anchor=(1, 0, 0.5, 1))

# Añadir un título al gráfico
plt.title('Distribución de Riesgo (solo M)')

# Mostrar el gráfico
plt.show()
```

```python
# Filtrar el DataFrame donde la columna C2 es 'F'
df_filtrado = tabla4[tabla4['8_Sexo'] == 'F']

# Contar las frecuencias de cada categoría en la columna C1 para el DataFrame filtrado
frecuencias = df_filtrado['CLASIFICACION FINAL DEL RIESGO'].value_counts()

# Crear el gráfico de torta
plt.figure(figsize=(8, 8))
wedges, texts, autotexts = plt.pie(
    frecuencias,
    labels=frecuencias.index,
    autopct='%1.1f%%',
    startangle=140,
    colors=plt.cm.Paired(range(len(frecuencias)))
)

# Añadir una leyenda
plt.legend(wedges, frecuencias.index, title="Riesgo", loc="center left", bbox_to_anchor=(1, 0, 0.5, 1))

# Añadir un título al gráfico
plt.title('Distribución de Riesgo (solo F)')

# Mostrar el gráfico
plt.show()
```
